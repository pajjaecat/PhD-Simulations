### Plot data for year 2010-2011

```julia
BaseHems.setPlotFonts(12,12,12)

fig,ax = subplots(4,3,figsize=(11,12),sharex=true,sharey=true);
mCmpt=1

for rowss in 1:4 
    
    for colls in 1:3
        maxx = round(maximum(PvStarYear2010_11[mCmpt]),digits=1)
        summ = round(sum(PvStarYear2010_11[mCmpt])*dt,digits=1)
        
        var1 = string(L"Max(P_v^*) =", maxx,"kW\n",
                     L" \sum P_v^* = ", summ,"kWh")
        var12 = string(L"Max(P_v^*) =", maxx,"kW ",
                     L" \sum P_v^* = ", summ,"kWh") 
        
        ax[rowss,colls].plot(PvStarYear2010_11[mCmpt]',color="k", lw=0.2,alpha=0.3)
        ax[rowss,colls].axhline(maximum(PvStarYear2010_11[mCmpt]),color="r",lw=0.5,alpha=1,ls=":")
        ax[rowss,colls].annotate(var1, (1, 1),color="red", xytext=(54,maximum(PvStarYear2010_11[mCmpt])-.5),
            textcoords="data", fontsize=10, horizontalalignment="right")
        
        ax[rowss,colls].set(title = month2010_11[mCmpt]);
        
        if colls==1 ax[rowss,colls].set(ylabel="P(kw)") end
        
        if rowss >= 4 
            ax[rowss,colls].set(xlabel="Hour(H)", xticks=0:8:44, xticklabels = 0:4:22, xlim = (-2,55))
        end
        mCmpt+=1;
    end
end
fig.subplots_adjust(wspace=0.05)
```
